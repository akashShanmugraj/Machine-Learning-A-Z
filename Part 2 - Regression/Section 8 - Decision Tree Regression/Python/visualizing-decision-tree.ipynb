```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor as d
from sklearn import datasets, tree
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
clf = d(random_state=1234)
model = clf.fit(X,y)
```

#### Texual Representation of Regression Model
This method is particularly useful when there is *a lack of userinterface* or if the developer wants to *export the model data* into a text file

```python
text_rep = tree.export_text(clf)
print(text_rep)
```

#### Plotting a tree (more like a mapping of the model)

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
```

#### Visualizing using the `dtreeviz` module

```python
from dtreeviz.trees import dtreeviz
viz = dtreeviz(clf, X, y, target_name="target", feature_names=iris.feature_names, class_names=list(iris.target_names))
viz
```
