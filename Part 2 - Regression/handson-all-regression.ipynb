```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing and Visualizing the Dataset

```python
dataset = pd.read_csv('sampleData.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values

plt.scatter(X,y,color = 'olive')
plt.plot(X,y, color = 'teal')
plt.xlabel('Independent Variable [f(x)]')
plt.ylabel('Dependent Variable [y]')
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
linearRegressor = LinearRegression()
linearRegressor.fit(X,y)

# Visualizing predictions of Linear Regression
plt.scatter(X,y, color = 'red')
plt.plot(X, linearRegressor.predict(X), color = 'brown')
plt.plot(X,y, color = 'yellow')
plt.title('Predictions using a Linear Regression\nRed scatter plot and Yellow plot lines represent actual dataset\nBrown plot lines represent predictions of the Linear Regressor')
plt.xlabel('Independent Variable [f(x)]')
plt.ylabel('Dependent Variable [y]')
```

```python
print(f'Coefficient is {linearRegressor.coef_} and Intercept is {linearRegressor.intercept_}')
```

Observe that the linear equation computed by the regressor is `y = 2.99105 * X + 0.15105` which is pretty close to our actual intent of our dataset, `y = 3 * x + 0`
The difference in the slope and intercept values can be attributed to our manual deviation in data points and errors of the linear regression model
